#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
from pandas_datareader import data as web
import yfinance as yf

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as np
```

```python
%%time
yf.pdr_override()

lista_ativos = ['TAEE11.SA', 'BBDC4.SA', 'KLBN3.SA']
dt_ini = '2021-10-02'

df = web.get_data_yahoo(lista_ativos,
                        start = dt_ini)


df.to_csv('ativos.csv')
df.head()
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
#criando um laço for para criação de variável 'Volatility' para os ativos selecionados:
for i in lista_ativos:
    df[('Volatility', i)] = df[('High', i)] - df[('Low', i)]
```

```python
#fazendo o stack dos ativos:
volatilidade_vertical = df['Volatility'].stack().reset_index()
#definindo o nome das colunas do stack:
volatilidade_vertical.columns = ['data', 'ativo', 'valor']
#exibindo o dataframe stackado:
volatilidade_vertical.head()
```

```python
#criando o gráfico do plotly para exibição da volatilidade dos ativos ao longo do tempo:
fig = px.line(volatilidade_vertical, x = 'data', y = 'valor', color = 'ativo', template = 'plotly_dark')
fig.show()
```

## 3. Resposta: O ativo com maior volatilidade parece ser o TAEE11.
