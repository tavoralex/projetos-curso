````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


prev = pd.read_csv('Profissão- Cientista de Dados_M10_support material.csv')
prev = prev.drop(columns=['Unnamed: 0', 'index', 'data_ref'])
prev.head()
````

````python
prev.info()
````

````python
veiculo = prev['posse_de_veiculo'].map({True:'possui', False:'nao possui'})
imovel = prev['posse_de_imovel'].map({True:'possui', False:'nao possui'})
````

````python
sns.pointplot(x=imovel, y='renda', data=prev, dodge=True, ci=95)
````

````python
sns.pointplot(x=veiculo, y='renda', data=prev, dodge=True, ci=95)
````

````python
#As duas me parecem ser estatisticamente relevantes, porém a posse de veículo parece se destacar em relação à diferença de média
# de rendas, onde quem possui veículo tem uma média de renda superior a 10% em comparação a quem não tem.
````
