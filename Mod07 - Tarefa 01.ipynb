#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
db.sexo = db.sexo.map({'F':1, 'M':0})
db.posse_de_veiculo = db.posse_de_veiculo.map({'Y':1, 'N':0})
db.posse_de_imovel = db.posse_de_imovel.map({'Y':1, 'N':0})

variaveis_necessarias = ['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

df_dummies = pd.get_dummies(db[variaveis_necessarias])
db_tratado = pd.concat([db, df_dummies], axis = 1)
db_tratado.drop(columns = variaveis_necessarias, inplace = True)
db_tratado.mau = db_tratado.mau.astype(int)
db_tratado
````
