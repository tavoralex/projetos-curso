````python
# 4) Apgar é uma nota que o pediatra dá ao bebê quando nasce de acordo com algumas
# características associadas principalmente à respiração. Apgar 1 e Apgar 5 são as notas 1 e 5 minutos
# do nascimento. Apgar5 será a nossa variável de interesse principal. Então remova todos os registros
# com Apgar5 não preenchido. Para esta seleção, conte novamente o número de linhas e o número de 
# missings.
# seu código aqui:

df_ex4 = dt_sinasc['APGAR5'].dropna()
print(df_ex4.count())
print(df_ex4.isna().sum())
````

````python
# 5) observe que as variáveis ['ESTCIVMAE', 'CONSULTAS'] possuem o código 9, que significa ignorado. 
# Vamos assumir que o não preenchido é o mesmo que o código 9.
# seu código aqui

df_ex5 = dt_sinasc.loc[:,['ESTCIVMAE', 'CONSULTAS']].fillna(9)
df_ex5.isna().sum()
````

````python
# 6) Substitua os valores faltantes da quantitativa (QTDFILVIVO) por zero.
# Seu código aqui

df_ex6 = dt_sinasc['QTDFILVIVO'].fillna(0)
df_ex6.isna().sum()
````

````python
# 7) Das restantes, decida que valore te parece mais adequado (um 'não preenchido' ou um valor 
# 'mais provável' como no item anterior) e preencha. Justifique. Lembre-se de que tratamento de 
# dados é trabalho do cientista, e que estamos tomando decisões a todo o momento - não há 
# necessariamente certo e errado aqui.
#seu código aqui

df_ex7 = dt_sinasc.loc[:,['LOCNASC', 'IDADEMAE', 'ESCMAE', 'GESTACAO', 'GRAVIDEZ']]
df_ex7.isna().sum()

#Entre as colunas acima temos valores não preenchidos apenas em 'ESCMAE', 'GESTACAO' e GRAVIDEZ,
#então, manterei apenas as mesmas em visualização por ora:

df_ex7.loc[:, ['ESCMAE', 'GESTACAO', 'GRAVIDEZ']]

df_ex7['ESCMAE'].value_counts()

#Avaliando que uma pequena parcela tem escolaridade declarada como nenhuma, e que os valores não
#preenchidos podem tanto ser devido a vergonha de preencher a questão quanto devido a não 
#entenderem a pergunta, considerarei os valores não preenchidos em 'ESCMAE' como 0.

df_ex7['ESCMAE'].fillna(0, inplace = True)

# Devido a não conseguir pensar em uma justificativa para os dados de 'GESTACAO' estarem ausentes ou
# serem devido as mães não saberem quando iniciaram a gestação ou a concluirem fora de ambiente
# hospitalar ou sem a devida assistência médica pré-natal, e apesar destes fatores a grande maioria 
# tender a ter uma gestação de tempo normal, distribuirei proporcionalmente com o método df.ffill().
# Preencherei também proporcionalmente a coluna 'GRAVIDEZ' por não conseguir pensar em uma
# justificativa para os dados não preenchidos.

df_ex7['GESTACAO'].ffill(inplace = True)
df_ex7['GRAVIDEZ'].ffill(inplace = True)


````

````python
# 8) Na questão 4 é citada a 'APGAR5' como variável de interesse principal, então a utilizarei para 
# resolver esta questão.
# seu código aqui

dt_sinasc.loc[dt_sinasc['APGAR5']>=8,'faixa_APGAR'] = 'normal'
dt_sinasc.loc[(dt_sinasc['APGAR5']>=6) & (dt_sinasc['APGAR5']<8),'faixa_APGAR'] = 'asfixia leve'
dt_sinasc.loc[(dt_sinasc['APGAR5']>=4) & (dt_sinasc['APGAR5']<6),'faixa_APGAR'] = 'asfixia moderada'
dt_sinasc.loc[(dt_sinasc['APGAR5']>= 0) & (dt_sinasc['APGAR5']<4),'faixa_APGAR'] = 'asfixia severa'
dt_sinasc['faixa_APGAR'].value_counts()
````
