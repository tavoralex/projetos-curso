```python
import pandas as pd
import numpy as np
```

```python
warnings.filterwarnings("ignore")

f = lambda x: x**3

def calcula_area(a, b, func, frame_num):
    num_bins=frame_num+1 # Quantidade de ret칙ngulos
    delta = (b-a)/num_bins # largura do ret칙ngulo
    x_cols = np.linspace(a, b-delta, num_bins)+delta/2 # Divis칚o do eixo X
    y_cols = f(x_cols) # alturas dos ret칙ngulos
    return sum(delta*y_cols)
    

def animate(n, a=0, b=np.pi, func=f, verdadeiro_valor = 2):
    
    plt.cla()  # Limpa o gr치fico anterior para montar o pr칩xim frame
    ax = plt.axes(xlim=(a, b), ylim = (0, 1.2))  # Monta o sistema de eixos
    
    # Plota a fun칞칚o
    x = np.linspace(a ,b, 100)
    y = f(x)
    ax.plot(x, y, '-');
    
    # Define as quebras e as alturas das colunas
    num_bins=n+1
    delta = (b-a)/num_bins
    x_cols = np.linspace(a, b-delta, num_bins)+delta/2
    y_cols = f(x_cols)

    # plota barras
    bar = ax.bar(x_cols, y_cols, alpha=.5, width = delta, edgecolor='blue')   

    # 츼rea dos ret칙ngulos
    soma_riemann = calcula_area(a, b, func, n)
    
    # Erro
    erro = verdadeiro_valor - soma_riemann
    ax.set_title(f'Soma de Riemann = {soma_riemann:.2f} erro = {erro:.4f}')

    return bar
animate(40);
```

### 1.  Foram utilizados 40 ret칙ngulos. A partir de 40, a soma de Rieman e o erro deixam de variar. Abaixo o c치lculo da fun칞칚o:

```python
print(f'Integral da fun칞칚o 洧논3 avaliada entre os pontos 1 e 2:   {calcula_area(1, 2, f, 40)}')
```
