Este modelo deve prever a variável ```renda``` com base nas demais variáveis exceto ```data_ref``` e ```index```. Já vimos que a variável renda é melhor modelada com a transformação ```log()```, não se esqueça disso.

````python
# Imports
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf


import scipy.stats as ss 
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

import patsy
import warnings

import statsmodels.api as sm

from scipy.interpolate import interp1d
````

````python
# Carregando dataset
df = pd.read_csv('previsao_de_renda_II.csv', index_col=0)
````

````python
# Amostra dos dados
df.head()
````

````python
# Info
df.info()
````

````python
# Preenche os valores missing pela média
df['tempo_emprego'].fillna(value=df['tempo_emprego'].mean(), inplace=True)
````

````python
# Separando bases de treino e teste
df_train = df.loc[:'2015-12-01'].copy()
df_test = df.loc['2016-01-01':].copy()

# Criando as variáveis log_renda
df_train['log_renda'] = np.log(df_train['renda'])
df_test['log_renda'] = np.log(df_test['renda'])
````

````python
X1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + qtd_filhos 
                    + idade
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(X1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````

````python
# Vamos remover mais uma variável do modelo já que o R² não se alterou
# A variável idade será removida, pois representa um coeficiente muito pequeno

X1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(X1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````
